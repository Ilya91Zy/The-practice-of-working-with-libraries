# График функции sin() на интервале от 0 до 3.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

fig, ax = plt.subplots()     #подключаем функцию subplots() для создания двух графиков в одной сетке  
plt.axis([-0.2, 3.2, -1.3, 1.3]) #задаем диапазон для осей х минимум, х максимум, у минимум, у максимум 
x = np.arange(0,3,0.1) # задаем значение интервала от 0 до 3 с шагом 0,1
y = np.sin(math.pi*x) # вызываем функцию синус х*ПИ
y1 = np.sin(math.pi*x +math.pi/2) #вызываем функцию синус х*ПИ + ПИ/2
y2 = np.sin(math.pi*x - math.pi/2)#вызываем функцию синус х*ПИ - ПИ/2
plt.title(r'Синусы', fontsize=16, color='blue')
plt.xlabel('ось X', fontsize=14) 
plt.ylabel('ось Y', fontsize=14) 
ax.plot(x, y, label = 'sin(Пx)', linewidth = 3, marker = '^', markersize=6, color = 'MediumVioletRed') #передача функции синус х*ПИ
ax.plot(x,y1, label = 'sinПХ + П/2', linewidth = 3, marker = 's', markersize = 6, color = 'navy') #передача функции функции синус х*ПИ + ПИ/2
ax.plot(x,y2, label = 'sin(Пx-П/2)', linewidth = 3, marker = 'x', markersize = 6, color = 'navy') #передача функции функции синус х*ПИ - ПИ/2
ax.legend()
fig.set_figwidth(10)
fig.set_figheight(6)
fig.set_facecolor('#FFE4B6') #заполнение цветом фигуры
ax.set_facecolor('#FFE4B5') #заполнение цветом 
plt.grid()
plt.show()
```

# График функции sin() на интервале от -7 до 7.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

fig, ax = plt.subplots()
plt.axis([-7.2,7.2,-1.5,5])
plt.title(r'Синусы', fontsize=16, color='blue')
x = np.arange(-7, 7, 0.1)
y = np.sin(x)
plt.xlabel('ось X', fontsize=14) 
plt.ylabel('ось Y', fontsize=14)
plt.grid()
ax.plot(x, y, linestyle = '-', linewidth = 1, color = 'Tomato')
ax.plot(x, y + 1, linestyle = '--', linewidth = 2, color = 'darkmagenta')
ax.plot(x, y + 2, linestyle = '-.', linewidth = 4, color = 'DimGrey')
ax.plot(x, y + 3, linestyle = ':', linewidth = 5, color = 'darkblue')
fig.set_figwidth(10)
fig.set_figheight(6)
fig.set_facecolor('#FFE4E1')
ax.set_facecolor('#FFE4E5')
plt.show()
```

# График 3D: «седло» F(x,y) = x**2 – y**2

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

x = np.arange(-6,6,0.3)
y = np.arange(-6,6,0.3)
x_1, y_1 = np.meshgrid(x,y)
z = x_1**2 - y_1**2
plt.figure(figsize=(10,8))
fig = plt.figure(figsize =(10 ,8))
axes = fig.add_subplot(projection='3d')
axes. plot_surface (x_1, y_1, z, cmap='magma', linewidth = 0.5,
                    rstride = 2, cstride = 2)
axes.set_xlabel ('$X$')
axes.set_ylabel ('$Y$')
axes.set_zlabel ('$Z$')
axes. set_title ('Седло')
fig.set_facecolor('palegreen')
axes.set_facecolor('palegreen')
plt.show ()
```
